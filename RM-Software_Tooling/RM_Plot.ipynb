```python
file=open("Rice_Gain.csv")

year=[]
rice=[]
jowar=[]
maize=[]
ragi=[]

count=0

for line in file.readlines() :
    
    if count != 0 :
        
        l=line.split(",")
        
        year.append(int(l[0].split("-")[0]))
        
        rice.append(int(l[1]))
        
        jowar.append(int(l[2]))
        
        maize.append(int(l[3]))
        
        ragi.append(int(l[4]))
        
    count+=1
    
print(year)
print(rice)
print(jowar)
print(maize)
print(ragi)
```

```python
import matplotlib.pyplot as plt
```

```python
data=[rice,jowar,maize,ragi]

fig = plt.figure(figsize =(10, 7))

ax = fig.add_subplot(111)

plot = ax.boxplot(data,patch_artist=True,vert=1)

colors = ['black','red', 'blue', 'green']
for patch,color in zip(plot['boxes'], colors):
    patch.set_facecolor(color)


#ax.boxplot(data)

ax.set_xticklabels(["rice","jowar","maize", "ragi"])

plt.title("Food Grains From Year 2001-2017")

plt.show()

data=[jowar,maize,ragi]

fig = plt.figure(figsize =(10, 7))

ax = fig.add_subplot(111)

plot = ax.boxplot(data,patch_artist=True,vert=1)

colors = ['red', 'blue', 'green']
for patch,color in zip(plot['boxes'], colors):
    patch.set_facecolor(color)


#ax.boxplot(data)

ax.set_xticklabels(["jowar","maize", "ragi"])

plt.title("Food Grains From Year 2001-2017")

plt.show()
    
```

```python
plt.scatter(jowar,maize,c="blue")

plt.xlabel("Jowar")
plt.ylabel("Maize")

plt.show()

plt.scatter(rice,ragi,c="red")

plt.xlabel("Rice")
plt.ylabel("Ragi")

plt.show()
```

```python
fig = plt.subplots(figsize =(12, 8))

barW=0.25

rice_p = [x for x in range(len(rice))]
jowar_p = [x+barW for x in rice_p ]
maize_p = [x+barW for x in jowar_p]
ragi_p = [x+barW for x in maize_p]

plt.bar(rice_p,rice,color ='black',width = barW,edgecolor ='grey',label ='Rice')
plt.bar(jowar_p,jowar,color ='r',width = barW,edgecolor ='grey',label ='Jowar')
plt.bar(maize_p,maize,color ='b',width = barW,edgecolor ='grey',label ='Maize')
plt.bar(ragi_p,ragi,color ='g',width = barW,edgecolor ='grey',label ='Ragi')

plt.xlabel("Year")
plt.ylabel("Production")
plt.xticks([x+barW for x in range(len(rice))],year)

plt.legend()
plt.show()

fig = plt.subplots(figsize =(12, 8))
barW=0.25

jowar_p = [x for x in range(len(rice))]
maize_p = [x+barW for x in jowar_p ]
ragi_p = [x+barW for x in maize_p]


plt.bar(jowar_p,jowar,color ='r',width = barW,edgecolor ='grey',label ='Jowar')
plt.bar(maize_p,maize,color ='b',width = barW,edgecolor ='grey',label ='Maize')
plt.bar(ragi_p,ragi,color ='g',width = barW,edgecolor ='grey',label ='Ragi')

plt.xlabel("Year")
plt.ylabel("Production")
plt.xticks([x+barW for x in range(len(rice))],year)

plt.legend()
plt.show()
```
